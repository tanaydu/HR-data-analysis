# Insights and Summary

 Note - All the Analysis is done on the assumption that this is the resignation data . Also this 
 data is representation of population data .
 
 1. The data contains 9 columns in which categorical has followimg uniques in them
     
    The company has 37 different business units .
    The company has 58 different designations in it.
    The company has 22 different Departments in it . 
    The company has 96 different sub departments in it.
    The company is based on 24 different locations.
    
    
    
 2. Feature Engineering or creating extra information from the existing data
     
     a) I created new column called 'Tenure in months' by difference of resigning and joining dates which tells us 
about the tenure served by a employee in the company in months.
        
     b) I created new column called 'Notice period in Days' which tells us that how much notice period is being 
served by a leaving employee in days by the difference of Last working day and Resignation date .
        
     c) I created new column called 'Zone' in which i binned the cities in 5 zones of india that is North , South 
West , East and Central India in order to see if a pattern is being followed in the cities only or 
also in a whole particular zone so extra information can be taken .
        
     d) I created a new column called "Month of joining' from date of joining column in which joining month of 
a employee is mentioned so that it can inferred that in which month the hiring is most in respect 
to location , department , business unit etc so it can be further used in analysis . 
        
     e) I created a new column called 'Month of Resign' from Resignation Date column in which resigning month of 
a employee is mentioned so that it can inferred that in which month the employees are resigning the most in           respect to location , department , business unit etc so it can help HR by determining that when hiring can
be done and prior arrangements can be made . 

        
        
 3. Individual analysis of Tenure in Months and Notice period in days 
     a) Tenure in Months :
         The average Tenure in our company in months is around 7.42 months.
         The maximum Tenure in our company is around 38.1 months .
         The minimum Tenure in our company is 0.0 months which means some employees had left company on the same 
         date of joining . 
         80% of the employees are having a tenure of only around 11 months or less than 11 months which can be                  seen in boxplot which is alarming as employees are not staying in the company for long .
         
      b) Notice period In days :
        The average  in our Notice period In days company in days is around 11 days.
        The maximum Notice period In days in our company is around 61 days .
        The minimum Notice period In days in our company is 0 days which means some employees had left company
        without even serving any notice period .
        80% of the employees are serving notice period of 28 days or less than 28 days which can be seen inboxplot.
        
        Also we found Significant Positive correlation between Tenure in months and Notice period in days that 
        means they both increase when either of one increases and both decrease when either of the one decreases .
